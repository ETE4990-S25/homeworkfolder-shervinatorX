```python
import json

# Define item class
class Item:
    def __init__(self, name, description, bonus_type, bonus_value):
        self.name = name
        self.description = description
        self.bonus_type = bonus_type
        self.bonus_value = bonus_value

    def get_details(self):
        return f"{self.name}: {self.description} | {self.bonus_type} +{self.bonus_value}"

# Define player class
class Player:
    def __init__(self, name, player_type):
        self.name = name
        self.player_type = player_type
        self.inventory = []
        self.stats = {"health": 100, "magic": 10 if player_type == "wizard" else 0, "strength": 10 if player_type == "warrior" else 5}

    def show_inventory(self):
        if not self.inventory:
            print("Inventory is empty.")
        else:
            print("\nInventory:")
            for item in self.inventory:
                print(f"- {item.get_details()}")

    def add_item(self, item):
        self.inventory.append(item)
        print(f"{item.name} added to inventory.")

    def remove_item(self, item_name):
        for item in self.inventory:
            if item.name == item_name:
                self.inventory.remove(item)
                print(f"{item.name} removed from inventory.")
                return
        print("Item not found.")

    def save_player(self):
        with open("player_data.json", "w") as file:
            json.dump(self.__dict__, file, default=lambda o: o.__dict__, indent=4)
        print("Game saved!")

    def load_player(self):
        try:
            with open("player_data.json", "r") as file:
                data = json.load(file)
                self.__dict__.update(data)
                self.inventory = [Item(**item) for item in self.inventory]
            print("Game loaded!")
        except FileNotFoundError:
            print("No saved data found.")

# Function to create a player
def create_player():
    name = input("Enter your character's name: ")
    player_type = input("Choose your class (wizard/warrior): ").lower()

    if player_type not in ["wizard", "warrior"]:
        print("Invalid choice. Defaulting to warrior.")
        player_type = "warrior"

    return Player(name, player_type)

# Initialize game
player = create_player()

# Predefined item list
item_list = [
    Item("Sword", "Forged in the depths of Polymar", "Strength", 5),
    Item("Staff", "Ancient staff with magical powers", "Magic", 10),
    Item("Shield", "Sturdy iron shield", "Defense", 7),
    Item("Potion", "Restores health", "Health", 20),
    Item("Dagger", "Small but deadly", "Strength", 3),
    Item("Ring of Fire", "Boosts fire magic", "Magic", 8),
    Item("Bow", "Silent but deadly", "Dexterity", 5),
    Item("Helmet", "Protects the head", "Defense", 4),
    Item("Boots", "Increases speed", "Agility", 2),
    Item("Amulet", "A mysterious charm", "Luck", 5)
]

# Add starter items
player.add_item(item_list[0])
player.add_item(item_list[3])

# Show inventory
player.show_inventory()

# Save player data
player.save_player()
```
