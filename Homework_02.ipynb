````python
# Base class for all animals
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print("Some generic animal sound")  # Generic placeholder, should be overridden

# Class for pets (subclass of Animal)
class Pet(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed
        self.friend = True
        self.eat = False  # Pets aren't sold for food

    def speak(self):
        print(f"{self.name} makes a friendly pet sound")

# Class for farm animals (subclass of Animal)
class FarmAnimal(Animal):
    def __init__(self, name, age, use_for_meat, use_for_wool=False, use_for_milk=False):
        super().__init__(name, age)
        self.use_for_meat = use_for_meat
        self.use_for_wool = use_for_wool
        self.use_for_milk = use_for_milk

    def speak(self):
        print(f"{self.name} makes a farm animal sound")

# Specific animal classes inheriting from FarmAnimal or Pet
class Dog(Pet):
    def __init__(self, name, age, breed="Unknown"):
        super().__init__(name, age, breed)
        self.guard_dog = False

    def speak(self):
        print(f"{self.name} barks!")

class Cow(FarmAnimal):
    def __init__(self, name, age):
        super().__init__(name, age, use_for_meat=True, use_for_milk=True)

    def speak(self):
        print(f"{self.name} says Moo!")

class Sheep(FarmAnimal):
    def __init__(self, name, age):
        super().__init__(name, age, use_for_meat=True, use_for_wool=True)

    def speak(self):
        print(f"{self.name} says Baa!")

# Function to manage animals on the farm
farm_animals = []

def add_animal(animal):
    farm_animals.append(animal)

def remove_animal(name):
    global farm_animals
    farm_animals = [animal for animal in farm_animals if animal.name != name]

def list_animals():
    if len(farm_animals) == 0:
        print("No animals on the farm!")
    for animal in farm_animals:
        print(f"Name: {animal.name}, Age: {animal.age}, Type: {type(animal).__name__}")

# Simulating user interaction
add_animal(Dog("Buddy", 4, "Labrador"))
add_animal(Cow("Betsy", 5))
add_animal(Sheep("Shaun", 2))

# Listing animals
list_animals()

# Making animals speak
for animal in farm_animals:
    animal.speak()

# Removing an animal
remove_animal("Buddy")

# Listing animals again
list_animals()
````
