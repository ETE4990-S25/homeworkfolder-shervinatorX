````python
# my project #1 Shervin Hosseinian
import json
import random

# Item Class
class Item:
    def __init__(self, name, category, attribute, description, stackable=False):
        self.name = name
        self.category = category
        self.attribute = attribute
        self.description = description
        self.stackable = stackable

    def to_dict(self):
        return {
            "item_name": self.name,
            "category": self.category,
            "attribute": self.attribute,
            "description": self.description,
            "stackable": self.stackable
        }

# Player Class
class Player:
    def __init__(self, name, role, skill, strength, powers, armor):
        self.name = name
        self.role = role
        self.skill = skill
        self.strength = strength
        self.powers = powers
        self.armor = armor
        self.inventory = []
        self.equipped_item = None

    def display_stats(self):
        print(f"\nPlayer: {self.name} ({self.role})")
        print(f"Skill: {self.skill}\nStrength: {self.strength}\nPowers: {self.powers}\nArmor: {self.armor}\n")

    def add_to_inventory(self, item):
        self.inventory.append(item)
        print(f"{item.name} added to inventory!")

    def show_inventory(self):
        if not self.inventory:
            print("Inventory is empty.")
        else:
            print("\nInventory:")
            for item in self.inventory:
                print(f"{item.name}: {item.attribute} - {item.description}")

    def equip_item(self, item_name):
        for item in self.inventory:
            if item.name == item_name:
                self.equipped_item = item
                print(f"You have equipped {item_name}.")
                return
        print("Item not found in inventory.")

    def unequip_item(self):
        if self.equipped_item:
            print(f"You have unequipped {self.equipped_item.name}.")
            self.equipped_item = None
        else:
            print("No item is currently equipped.")

# Save and Load Functions
def save_game(player):
    data = {
        'name': player.name,
        'role': player.role,
        'skill': player.skill,
        'strength': player.strength,
        'powers': player.powers,
        'armor': player.armor,
        'inventory': [item.to_dict() for item in player.inventory],
        'equipped_item': player.equipped_item.to_dict() if player.equipped_item else None
    }
    with open('saved_game.json', 'w') as f:
        json.dump(data, f)
    print("Game saved successfully!")

def load_game():
    try:
        with open('saved_game.json', 'r') as f:
            data = json.load(f)
        player = Player(data['name'], data['role'], data['skill'], data['strength'], data['powers'], data['armor'])
        player.inventory = [Item(**item) for item in data['inventory']]
        player.equipped_item = Item(**data['equipped_item']) if data.get('equipped_item') else None
        print("Game loaded successfully!")
        return player
    except FileNotFoundError:
        print("No saved game found. Starting new game...")
        return None

# Load Inventory Data
def load_inventory(filename):
    try:
        with open(filename, 'r') as f:
            items = json.load(f)
            return [Item(**item) for item in items]
    except FileNotFoundError:
        print(f"{filename} not found. Using default empty inventory.")
        return []

inventories = {
    "Hero": load_inventory("inventoryhero.json"),
    "Villain": load_inventory("inventoryvillain.json"),
    "Civilian": load_inventory("inventorycivilian.json")
}

# Simple Combat System
def combat(player):
    enemy_health = 20
    print("\nA wild enemy appears!")
    while enemy_health > 0:
        attack = random.randint(1, player.strength + player.powers)
        enemy_health -= attack
        print(f"You attack the enemy for {attack} damage! Enemy health: {max(0, enemy_health)}")
        if enemy_health <= 0:
            print("You defeated the enemy!")
            break
        enemy_attack = random.randint(1, 5)
        print(f"The enemy attacks you for {enemy_attack} damage!")

# Game Setup
roles = {
    "Hero": Player("", "Hero", 5, 7, 8, 5),
    "Villain": Player("", "Villain", 8, 5, 7, 6),
    "Civilian": Player("", "Civilian", 5, 2, 0, 0)
}

print("Welcome to the RPG Adventure!")
load_choice = input("Would you like to load a saved game? (yes/no): ").strip().lower()
player = load_game() if load_choice == "yes" else None

if not player:
    name = input("Enter your name: ")
    print(f"You entered: {name}")
    while True:
        role_choice = input("Choose your role (Hero, Villain, Civilian): ").capitalize()
        if role_choice in roles:
            player = roles[role_choice]
            player.name = name
            break
        else:
            print("Invalid choice. Try again.")

player.display_stats()

print("\nChoose your starting item:")
for i, item in enumerate(inventories[player.role], 1):
    print(f"{i}. {item.name}: {item.attribute}")

while True:
    try:
        choice = int(input("Enter the item number: ")) - 1
        if 0 <= choice < len(inventories[player.role]):
            player.add_to_inventory(inventories[player.role][choice])
            break
        else:
            print("Invalid choice, try again.")
    except ValueError:
        print("Please enter a valid number.")

player.show_inventory()

combat(player)
save_game(player)
print("Game Over.")
````
